```python


import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


```

```python
# Example true and predicted values
y_true = np.array([3.0, -0.5, 2.0, 7.0])
y_pred = np.array([2.5, 0.0, 2.1, 7.8])




```

Given:
**True (y)**=[3.0,−0.5,2.0,7.0]
**Predicted(y)**=[2.5,0.0,2.1,7.8]

| i | True (y) | Predicted ($\hat{y}$) | Error (y - $\hat{y}$) | Squared Error |
| - | -------- | --------------------- | --------------------- | ------------- |
| 1 | 3.0      | 2.5                   | 0.5                   | 0.25          |
| 2 | -0.5     | 0.0                   | -0.5                  | 0.25          |
| 3 | 2.0      | 2.1                   | -0.1                  | 0.01          |
| 4 | 7.0      | 7.8                   | -0.8                  | 0.64          |

**Sum of squared errors**:0.25+0.25+0.01+0.64=1.15
**MSE**= 1.15/4=0.2875
**RMSE**=sqrt(MSE)=sqrt(0.28750)≈0.5367

## Mean Absolute error

| i | Error | Absolute Error |
| - | ----- | -------------- |
| 1 | 0.5   | 0.5            |
| 2 | -0.5  | 0.5            |
| 3 | -0.1  | 0.1            |
| 4 | -0.8  | 0.8            |

Sum of absolute errors:0.5+0.5+0.1+0.8=1.9
**MAE:**=1.9/4=0.475


## R-squared (R²) Calculation

**Formula:**

R^2 = 1 - SSres/SStotal




**Residual Sum of Squares (SS_res):**

SSres=∑(y− y^)^2


**Total Sum of Squares (SS_tot):**
SStotal=∑(y− y^)^2



**Mean of \(y\):**

ȳ = (3.0 + (-0.5) + 2.0 + 7.0) / 4 = 2.875



| i | y   | y - mean | (y - mean)² |
|---|-----|-----------|--------------|
| 1 | 3.0 | 0.125     | 0.015625     |
| 2 | -0.5 | -3.375   | 11.390625    |
| 3 | 2.0 | -0.875    | 0.765625     |
| 4 | 7.0 | 4.125     | 17.015625    |

##### Total sum of squares:
SS_tot = (3.0 - 2.875)² + (-0.5 - 2.875)² + (2.0 - 2.875)² + (7.0 - 2.875)²  
       = 0.015625 + 11.390625 + 0.765625 + 17.015625  
       = 29.1875

#### R²:
R² = 1 - (SS_res / SS_tot)  
   = 1 - (1.15 / 29.1875)  
   ≈ 1 - 0.0394 = 0.9606

## Note:

R² close to 1: Good fit

R² close to 0: Poor fit


```python
# Calculate MSE
mse = mean_squared_error(y_true, y_pred)
```

```python

# Calculate RMSE
rmse = np.sqrt(mse)

```

```python
# Calculate MAE
mae = mean_absolute_error(y_true, y_pred)

```

```python
# Calculate R2 Score
r2 = r2_score(y_true, y_pred)
```

```python
print(f"Mean Squared Error (MSE): {mse:.3f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.3f}")
print(f"Mean Absolute Error (MAE): {mae:.3f}")
print(f"R-squared (R²): {r2:.3f}")
```

```python
# Plot true vs predicted
plt.scatter(y_true, y_pred, color='blue')
plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], color='red', linestyle='--')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('True vs Predicted')
plt.show()
```
